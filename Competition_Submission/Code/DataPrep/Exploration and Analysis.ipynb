## Check Log Price Distribution

```python
# Analize pricing distribution

# Histogram
print(df_train['log_price'].describe().round())
plt.hist(df_train['log_price'], bins=100)
plt.show()
# Box plot
plt.boxplot(df_train['log_price'])
plt.xticks([1], ['log_price'], rotation='horizontal')
plt.show()
```

## Check Correlation

```python
list_of_standard_columns = ["log_price","accommodates","bathrooms","cancellation_policy","cleaning_fee",
                            "first_review","host_has_profile_pic","host_identity_verified","host_response_rate",
                            "host_since","instant_bookable","last_review","latitude","longitude", "neighbourhood",
                            "number_of_reviews","review_scores_rating","thumbnail_url","bedrooms","beds"]
# heatmap
df_heatMap = df_train[list_of_standard_columns]
corrmat = df_heatMap.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True)
plt.show()
```

```python
def plotfunction(x,y):
    plt.figure(figsize=(10,10))
    plt.xlabel(x)
    plt.ylabel(y)
    plt.plot(df_train[x], df_train[y], 'o', alpha=0.01)
    plt.show()
    
```

## Plot Columns

```python
for column in list_of_standard_columns:
    plotfunction('log_price', column)
```
