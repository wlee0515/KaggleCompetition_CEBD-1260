### Amenities

```python
list_amenities = []

for index, row in df_train.iterrows():
    local_list = row["amenities"].strip("{").strip("}").split(",")
    for item in local_list:
        if "" == item:
            continue
        if not item in list_amenities:
            if -1 == item.find("translation missing"):
                list_amenities.append(item)

for index, row in df_test.iterrows():
    local_list = row["amenities"].strip("{").strip("}").split(",")
    for item in local_list:
        if "" == item:
            continue
        if not item in list_amenities:
            if -1 == item.find("translation missing"):
                list_amenities.append(item)

print(list_amenities)
print(len(list_amenities))
```

```python
for item in list_amenities:
    df_train_new[item] = 0
    df_test_new[item] = 0

    
for index, row in df_train_new.iterrows():
    local_list = row["amenities"].strip("{").strip("}").split(",")
    for item in local_list:
        if item in list_amenities:
            row[item] = 1
                
for index, row in df_test_new.iterrows():
    local_list = row["amenities"].strip("{").strip("}").split(",")
    for item in local_list:
        if item in list_amenities:
            row[item] = 1


df_train_new = df_train_new.drop("amenities", axis=1)
df_test_new = df_test_new.drop("amenities", axis=1)
    
```

### Cancellation Policy

```python
list_cancellation_policy = df_train["cancellation_policy"].unique()
print(list_cancellation_policy)
list_cancellation_policy = df_test["cancellation_policy"].unique()
print(list_cancellation_policy)
```

```python
def cancelationfee_rating(rating):
    if rating == "flexible":
        return 1
    if rating == "moderate":
        return 2
    if rating == "strict":
        return 3
    if rating == "super_strict_30":
        return 4
    if rating == "super_strict_60":
        return 5
    return 0
```

```python
df_train_new["cancellation_policy"]=df_train_new["cancellation_policy"].apply(cancelationfee_rating)
list_cancellation_policy = df_train_new["cancellation_policy"].unique()
print(list_cancellation_policy)
```
